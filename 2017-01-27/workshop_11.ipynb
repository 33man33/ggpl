# Positioning functions

#### It was developed a function for objects positioning that takes as input an object and returns the same object at a desired point on the map, by translating and rotating it suitably.

```python
def place(r, model,X, Y):
	h = R([1,2])(rotate)(model)
	if r == PI:
		h = T([1,2])([SIZE([1])(h)[0], SIZE([2])(h)[0]])(h)
	elif r == PI/2:
		h = T([1])(SIZE([1])(h)[0])(h)
	elif r == PI/4:
		h = T([1])(SIZE([1])(h)[0]/2)(h)
	elif r == -PI/2:
		h = T([2])(SIZE([2])(h)[0])(h)
	h = T([1,2])([X,Y])(h)
	return h
```
